```python
STUDENT_ID = "TUPM-25-4494"
ENGINEER = "MATIC"

SEED_NUM = int(STUDENT_ID[-1])

print("SEED_NUM: ", SEED_NUM)

# [CELL 3] ASCII INSPECTION
first_letter = ENGINEER[0]
ASCII = ord(first_letter) # Convert Char to Integer Memory Code

print("ASCII: ", ASCII)

loops = SEED_NUM + 3

verification_hash = (ASCII * SEED_NUM) - loops

print("Verification_Hash: ", verification_hash)
```
